# What holiday do New Yorkers complain about the most?

In this notebook we investigate if the 4th of July is the loudest holiday of the year. To do this we are going to look at [311 noise complaints](https://nycopendata.socrata.com/Social-Services/311-Service-Requests-from-2010-to-Present/erm2-nwe9) from the New York open data portal. 

First we import the necessary libraries we are going to need.

There is a lot of rich data in there but we won’t need it all. Instead lets limit the columns we are bringing in to 
- Complaint Type

While not strictly necessary here, let's also load the data in chunks. This can be helpful if we want to do any filtering as we go,for example here, dropping anything without a descriptor or a complaint type. When reading in big files using pandas, this lets us avoid blowing up our memory requirements.


````python
cols = ['Complaint Type', 'Created Date','Descriptor','Borough','Latitude','Longitude']
````

````python
all_data = pd.DataFrame()
for index, chunk in enumerate(pd.read_csv('311_calls.csv', usecols=cols, chunksize=100000)):
    chunk  = chunk.dropna(subset=['Descriptor', 'Complaint Type'], how='all')
    all_data = all_data.append(chunk)
    
````

````python
all_data.shape
````

This will take a few moments to read in but then we should have all the 311 data we will need. 

We have over 21 million calls in there! New Yorkers complain a lot! 

Let’s look at the types of complaints that tend to come up.

````python
all_data['Complaint Type'].value_counts().head(20)
````

````python
all_data[all_data['Complaint Type'].str.contains("Noise")]['Complaint Type'].value_counts()
````

As we can see there are quite a few top level complaint types that indicate noise. Let's dig in to one in more detail, specifically the "Noise - Commercial" type 

````python
all_data[all_data['Complaint Type']=='Noise - Commercial'].Descriptor.value_counts()
````

While some of these are relevant to our question, some like 'Banging/Pounding" probably aren't. So let's narrow it down to calls with a complaint type which is one of 

- 'Noise - Residential', 
- 'Noise Survey',
- 'Noise - Park',
- 'Noise - Street/Sidewalk'

or has a descriptor of 

- 'Car/Truck Music'
- 'Car/Truck Horn'
- 'Loud Music/Party'

````python
cleaned = all_data[
    all_data['Complaint Type'].isin([
        'Noise - Residential', 
        'Noise Survey',
        'Noise - Park',
        'Noise - Street/Sidewalk'
    ]) | 
    all_data['Descriptor'].isin(['Car/Truck Music','Car/Truck Horn','Loud Music/Party']) 
]
````

````python
cleaned.shape
````

That still leaves us with 2,573,188 calls! Next let's format the date to be a proper python datetime

````python
cleaned = cleaned.assign(date = pd.to_datetime(cleaned['Created Date'], format='%m/%d/%Y %H:%M:%S %p'))
````

Now that we have the data in a good format, we can plot a time series the number of calls per day to see if we can spot any trends around the holidays. We will add a red line indicating the 4th of July each year 


````python
trend = cleaned.groupby(cleaned.date.dt.floor('d')).count()[['Descriptor']]

plt.figure(figsize=(15,5))
plt.scatter(trend.index,trend.Descriptor)
for year in range(2010,2020):
    plt.axvline(datetime(year,7,4), color='red')
plt.xlabel("Day")
plt.ylabel('No calls')
````

There seems to be a few things going on here that make it hard to pick out what is going on. 
- First we see that there is a general trend to an increased number of calls each year. 
- There tends to be more calls in the summer 
- It looks like there are two trends superimposed on each other

This all makes it hard to figure out what is going on and makes it really hard to tell if there is much of a difference on the 4th of July or not. 

We might guess that there is a different behaviour on the weekends and weekdays. Let's plot the same chart again but plot weekdays with blue and weekends with green

````python
trend = cleaned.groupby(cleaned.date.dt.floor('d')).count()[['Descriptor']]
weekdays = (pd.DatetimeIndex(trend.index).dayofweek) // 5 == 1

plt.figure(figsize=(15,5))
plt.scatter(trend[weekdays].index,trend[weekdays].Descriptor,color='blue',label='weekdays')
plt.scatter(trend[~weekdays].index,trend[~weekdays].Descriptor,color='green', label='weekends')
for year in range(2010,2020):
    plt.axvline(datetime(year,7,4), color='red')
plt.xlabel('day')
plt.ylabel('Number of Noise based 311 calls')
plt.legend()
````

It's pretty obvious that the is significantly different behaviour on the weekends as opposed to the weekdays. 

If the 4th of July happened to fall on a weekend one year, we wouldn't be able to tell easily if the number of calls are higher because it's the 4th or because it's a weekend. We need a way to separate out the different components of the number of calls.

To do that we need to build an additive model with components for 

- The day of the year 
- The overall trend 
- The day of the week 
- If there is a holiday on that day 

Thankfully there is a pretty good library from facebook that will help us with this called Prophet. 

Let's load it up and make a quick model with those components

````python
m = Prophet()
m.add_country_holidays(country_name='US')
m.fit(trend.reset_index().rename(columns={'date':'ds','Descriptor':'y'}))
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
````

The above code adds features to the model for various US holidays and then tries to fit that model to our data. We can plot the predictions of the model and compare with our data as follows:

````python
plt.figure(figsize=(15,5))
plt.scatter(trend.index,trend.Descriptor, alpha=1, label='data',color='black')
plt.fill_between(forecast.ds, forecast.yhat_lower, forecast.yhat_upper, alpha=0.2, color='blue',label='uncertanty')

plt.plot(forecast.ds,forecast.yhat,alpha=0.8, label='forecast',color='orange')
plt.xlim(datetime(2010,1,1),datetime(2019,1,1))
plt.legend();
````

The model does ok at capturing the trends we see but could probably be improved. We are just doing this for fun just now so let's leave tuning the model for another time and simply look at the different components of the fit model

````python
m.plot_components(forecast);
````

We see some interesting features here

- The model confirms our intuition that the weekends tend to have a larger volume of calls than the weekdays, in fact the weekends tend to have roughly 500 more calls than on average while weekdays tend to have 200 fewer calls than on average 
- There is a huge seasionality to the calls with summer having a lot more calls while the winter months have fewer 
- There are two peaks during the year for calls, one is around mid may and the latter around the start of september 
- We see some significant effects from the holidays

## By borough

We can also break this out by borough to see what kind of trends we see in each part of the city. We simple slice the data by borough and fit a model to each.

````python
trend_boroughs = cleaned.groupby([cleaned.date.dt.floor('d'),'Borough']).count()[['Descriptor']].reset_index()
````

````python
trend_boroughs.head()
````

````python
forecasts = {}
models = {}
for borough, data in trend_boroughs.groupby('Borough'):
    m = Prophet()
    m.add_country_holidays(country_name='US')
    m.fit(data.rename(columns={'date':'ds','Descriptor':'y'}))
    future = m.make_future_dataframe(periods=365)
    models[borough] = m 
    forecasts[borough] = m.predict(future)
````

Now we can compare the different model components with each other to see how the boroughs differ.

````python
all_boroughs  = pd.DataFrame()
for borough in models.keys():
    f= forecasts[borough]
    all_boroughs[borough] = f[f[holidays].abs().any(axis=1) > 0][holidays].drop_duplicates().max()
````

````python
all_boroughs.T.plot(kind='bar',figsize=(15,5))
````


It looks like the following are the most complained about holidays for each borough:

- Manhattan: Memorial Day, New Years Day
- Brooklyn: Labor Day and Memorial Day
- Bonx: Memorial day and Labor Day
- Queens: Independence Day and Memorial Day 
- Staten Island: Independence Day and Memorial day


````python
from fbprophet.plot import plot_weekly,plot_yearly
````

````python
fig, ax = plt.subplots(nrows=3,ncols=1,figsize=(10,12))

colors={
    "MANHATTAN" : 'red',
    "BRONX": 'blue',
    'QUEENS': 'purple',
    'BROOKLYN': 'green',
    'STATEN ISLAND': 'orange',
    'Unspecified': 'gray'
}
for borough in models.keys():
    forecast_borough = forecasts[borough]
    model_borough = models[borough]
    ax[0].plot(forecast_borough.ds, forecast_borough.trend, label=borough,color=colors[borough])
    ax[0].legend()
    
    a=plot_weekly(model_borough, ax=ax[1])
    a[0].set_color(colors[borough])
    a= plot_yearly(model_borough, ax=ax[2])
    a[0].set_color(colors[borough])
    
ax[0].set_ylabel('Trend')
ax[0].set_xlabel("Day")
````

It's interesting to see that the weekly trends are pretty similar but the seasonal trends are a little different for each borough.
